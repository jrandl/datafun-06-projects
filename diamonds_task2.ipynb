## Section 5-Unique: Use the Series method unique() to get unique category values

```python
import pandas as pd

# Read the 'diamonds.csv' file into a DataFrame
diamonds = pd.read_csv('diamonds.csv', index_col=0)

# Set the display precision to 2 decimal places
pd.set_option('display.precision', 2)

# Find the unique values in the category values
unique_values_cut = diamonds['cut'].unique()
unique_values_color = diamonds['color'].unique()
unique_values_clarity = diamonds['clarity'].unique()

# Display the unique values
print(unique_values_cut)
print(unique_values_color)
print(unique_values_clarity)
```

## Section 6-Histograms: Use the DataFrame's hist() function to create a histogram for each numerical column

```python
import pandas as pd
import matplotlib.pyplot as plt
import chardet

# Detect the encoding of the CSV file
with open('diamonds.csv', 'rb') as f:
    result = chardet.detect(f.read())

diamonds = pd.read_csv('diamonds.csv', encoding=result['encoding'])

# Set the display precision to 2 decimal places
pd.set_option('display.precision', 2)

columns_of_interest = [
    'carat', 'depth', 'table', 'price', 'x', 'y', 'z'
]

# Loop through each column and plot individual histograms
for col in columns_of_interest:
    plt.figure(figsize=(10, 5))  # Set the figure size for each plot
    diamonds[col].hist(bins=20)  # Adjust the number of bins if needed
    plt.title(f'Histogram of {col}')  # Set the title
    plt.xlabel(col)  # Set x-axis label
    plt.ylabel('Frequency')  # Set y-axis label
    plt.tight_layout()
    plt.show()
```
